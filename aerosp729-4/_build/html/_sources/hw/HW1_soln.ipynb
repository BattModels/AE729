```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
import niceplots

plt.style.use(niceplots.get_style())
colors = niceplots.get_colors_list()
```

```python
def OCV(Q , Q_max=5.080457832540185):
    coeffs = [-0.44934919, -0.07263436,  0.03009002, -0.02263346, -0.24777699, -0.26923143, 3.72029854]
    x = Q / Q_max
    result = 0.0
    for i in range(0, len(coeffs)-1):
        result += (coeffs[i] ) * ((2*x -1)**(i+1) - (2*x*i*(1-x))/(2*x-1)**(1-i))  # Higher-order terms

    result += coeffs[len(coeffs)-1] + (8.3145 * 298.15 / 96485.3321) * np.log((1-x)/x)
    return result
```

# Equivalent Circuit: Internal Resistor Model

```python
C_rates = [0.1, 0.5, 1.0]
# fig, ax = plt.subplots(1, 3, figsize=(12, 4))

avg_res_list = np.zeros(3)

for c_idx, data in enumerate(zip(currents, capacities, measured_voltages)):
    current, capacity, voltage = data
    
    capacity = capacity[1:-1]
    voltage = voltage[1:-1]

    v_loss = OCV(capacity) - voltage
    
    resistance = v_loss / current
    average_resistance = integrate.simpson(resistance, capacity) / (capacity[-1] - capacity[0])
    print(f"Average resistance value for {current} A case: {average_resistance} Ohms")

    avg_res_list[c_idx] = average_resistance
    
    plt.plot(capacity, v_loss, label=f"{current} A")
    plt.xlabel("Capacity")
    plt.ylabel("Overpotential [V]")
    plt.legend()
    
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

for c_idx, data in enumerate(zip(currents, capacities, measured_voltages)):
    current, capacity, voltage = data
    
    capacity = capacity[1:-1]
    voltage = voltage[1:-1]

    v_model = OCV(capacity) - current * avg_res_list[1]

    MAE = np.mean(np.abs(voltage - v_model))
    print(f"MAE for {current} A case: {MAE} V")
    
    ax[c_idx].plot(capacity, voltage, label="Measured")
    ax[c_idx].plot(capacity, v_model, label="Model")
    ax[c_idx].set_xlabel("Capacity")
    ax[c_idx].set_ylabel("Overpotential [V]")
    ax[c_idx].set_title(f"{current} A")
    ax[c_idx].legend()
```

# Equivalent Circuit: Resistor-Capacitor (RC) Model

```python
df = pd.read_csv("pulse.csv")
time_pulse = df["time_s"].to_numpy()
voltage_pulse = df["voltage_V"].to_numpy()
current_pulse = df["current_A"].to_numpy()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 5))
ax[0].plot(time_pulse, current_pulse)
ax[1].plot(time_pulse, voltage_pulse)
```

```python
idx = np.where(current_pulse == 0)[0]
```

```python
V_min = voltage_pulse[idx[0]-1]
Rint = (voltage_pulse[idx[0]] - V_min) / current_pulse[0]
print(f"The value of R_int is {Rint} Ohms")
```

```python
Q_discharge = 2.5 * 30 / 3600 # Ah
Rct = ((OCV(Q_discharge) - V_min) / current_pulse[0]) - Rint
print(f"The value of Rct is {Rct} Ohms")
```

```python
voltage_shift = voltage_pulse[idx] - voltage_pulse[idx[0]]
time_shift = time_pulse[idx] - time_pulse[idx[0]]

voltage_normalized = voltage_shift / (2.5 * Rct)
plt.plot(time_shift, voltage_normalized)
```

```python
tau_idx = np.where(voltage_normalized > 0.632)[0][0]
```

```python
print(voltage_normalized[tau_idx], voltage_normalized[tau_idx-1])
```

```python
tau = time_shift[tau_idx-1]
Cdl = tau / Rct
print(f"The time constant is {tau} s")
print(f"The double layer capacitance is {Cdl} F")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

for c_idx, data in enumerate(zip(currents, capacities, measured_voltages)):
    current, capacity, voltage = data
    
    capacity = capacity[1:-1]
    voltage = voltage[1:-1]

    time = capacity * 3600 / current
    v_model = OCV(capacity) - current * Rint - current * Rct * (1 - np.exp(-time/tau))

    MAE = np.mean(np.abs(voltage - v_model))
    print(f"MAE for {current} A case: {MAE} V")
    
    ax[c_idx].plot(capacity, voltage, label="Measured")
    ax[c_idx].plot(capacity, v_model, label="Model")
    ax[c_idx].set_xlabel("Capacity")
    ax[c_idx].set_ylabel("Overpotential [V]")
    ax[c_idx].set_title(f"{current} A")
    ax[c_idx].legend()
```
